```python
import pandas as pd
import matplotlib.pyplot as py
ch=1
while(ch!=0):
    print("********----------WELCOME TO THE FLIGHT BOOKING HOME PAGE----------*********")
    print("0, To Exit")
    print("1, Customer")
    print("2, Flight")
    print("3, Booking")
    ch=int(input("Enter your choice->"))
    if(ch==1):
        x=1
        while(x!=0):
            l=pd.read_csv("E:\csv\\cstmr.csv",index_col="CustomerID")
            print("----------------------------------------------------------------")
            print("1, To Display Data")
            print("2, To Add new customer")
            print("3, To Update customer details")
            print("4, To Remove customer details")
            print("0, To Back to home page")
            x=int(input("Enter your choice-"))
            if(x==1):
                print("----------------------------------------------------------------")
                print(l)
            elif(x==2):
                print("----------------------------------------------------------------")
                CID=int(input("Enter CustomerID want to add-"))
                if CID in l.index:
                    print("ID already exist")
                else:
                    NM=input("Enter the name-")
                    ADRS=input("Enter the address-")
                    PH=int(input("Enter the phone number-"))
                    l.loc[CID]=[NM,ADRS,PH]
                    print("\n Data added successfully \n")
                    print(l)
            elif(x==3):
                print("---------------------------------------------------------------")
                CID=int(input("Enter Custmer ID want to update-"))
                if CID in l.index:
                    NM=input("Enter the name-")
                    ADRS=input("Enter the address-")
                    PH=int(input("Enter the phone number-"))
                    l.loc[CID]=[NM,ADRS,PH]
                    print("\n Data updated successfully \n")
                    print(l)
                else:
                    print("\n ID doesn't exist \n")
                    
            elif(x==4):
                print("----------------------------------------------------------------")
                val=int(input("Enter the Customer ID,whose data you want to delete-"))
                if val in l.index:
                    l=l.drop(val)
                    print("\n Data removed successfully \n")
                    print(l)
                else:
                    print("\n ID doesn't exist \n")
               
            l.to_csv("E:\csv\\cstmr.csv")
    
    elif(ch==2):
        x=1
        while(x!=0):
            a=pd.read_csv("E:\csv\\flght.csv",index_col="FlightNo")
            print("\n***------------------------------------------------------***\n \n ")
            print("1, To show flight data")
            print("2, To add a new flight")
            print("3, To update flight details")
            print("4, To remove flight details")
            print("0, To Back to home page")
            x=int(input("enter your choice"))
            if(x==1):
                print("----------------------------------------------------------------")
                print(a)
            elif(x==2):
                print("----------------------------------------------------------------")
                FN=input("Enter flight number want to add-")
                if FN in a.index:
                    print("\nflight number already exist \n")
                else:
                    DT=input("Enter the date-")
                    DS=input("Enter the destination-")
                    AL=input("Enter the Airline-")
                    a.loc[FN]=[DT,DS,AL]
                    print("Data added successfully \n")
                    print(a)
            elif(x==3):
                print("----------------------------------------------------------------")
                FN=input("Enter flight number want to update-")
                if FN in a.index:
                    DT=input("Enter the date-")
                    DS=input("Enter the destination-")
                    AL=input("Enter the Airline-")
                    a.loc[FN]=[DT,DS,AL]
                    print("\n Data updated successfully \n")
                    print(a)
                else:
                    print("\n flight doesn't exist \n")
            elif(x==4):
                print("----------------------------------------------------------------")
                val=input("Enter the flight no.,whose details you want to delete-")
                if val in a.index:
                    a=a.drop(val)
                    print("\n Data removed successfully \n")
                    print(a)
                else:
                    print("\n flight doesn't exist \n")

            a.to_csv("E:\csv\\flght.csv")
    
    elif(ch==3):
        x=1
        while(x!=0):
            b=pd.read_csv("E:\csv\\bkng.csv",index_col="FlightNo")
            print("----------------------------------------------------------------")
            print("1, To display flight bookings")
            print("2, To add new flight booking")
            print("3, To cancel booking")
            print("4, To show the graph of booking fare")
            print("0, To Back to home page")
            x=int(input("Enter your choice-"))
            if(x==1):
                print("--------------------------------------------------------")
                print(b)
            elif(x==2):
                print("--------------------------------------------------------")
                AID=input("Enter Airlines ID-")
                if AID in b.index:
                    print("\n Data Already exists \n")
                else:
                    CID=input("Enter customer ID-")
                    NM=input("Enter the name-")
                    DT=input("Enter Date-")
                    CS=int(input("Enter Cost-"))
                    b.loc[AID]=[CID,NM,DT,CS]
                    print("\n Booking added successfully \n")
                    print(b)
            elif(x==3):
                print("--------------------------------------------------------")
                val=input("Enter the Airlines ID,whose booking you want to cancel-")
                if val in b.index:
                    b=b.drop(val)
                    print("\n Booking canceled successfully \n")
                    print(b)
                else:
                    print("\n Data doesn't exists \n")
            elif(x==4):
                py.figure(figsize=(15,7))
                py.bar(b["Name"],b["cost"])
                py.title("Booking fare details")
                py.xlabel("Customer name")
                py.ylabel("Amount in rs.")
                py.show()
            
            else:
                print("Invalid option")
            
            b.to_csv("E:\csv\\bkng.csv")
```
